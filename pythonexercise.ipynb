```python
# Exercise 35

data = df[['TOTPOPULAT', 'SCHTOT']].copy()

data = data.apply(pd.to_numeric, errors='coerce')

data = data.dropna()

x = data['TOTPOPULAT']
y = data['SCHTOT']

m, c = np.polyfit(x, y, 1)

line = np.poly1d((m, c))

plt.figure(figsize=(8,6))

plt.scatter(x, y)

plt.plot(x, line(x), color='red')

plt.xlabel("Total Population")
plt.ylabel("Total Schools")
plt.title("Population vs Schools")

plt.show()

print("Slope :", round(m, 4))
print("Intercept :", round(c, 4))
```

```python
# Exercise 36

cols = [
    'OVERALL_LI',
    'INFRA_SCORE',
    'SEXRATIO',
    'TCHTOT'
]

norm = pd.DataFrame()

for col in cols:
    norm[col] = (
        (df[col]-df[col].min()) /
        (df[col].max()-df[col].min())
    )

norm['TCHTOT'] = 1 - norm['TCHTOT']

df['EDI'] = norm.mean(axis=1)

display(
    df.sort_values('EDI', ascending=False)
      [['STATNAME','DISTNAME','EDI']]
      .head(10)
)

display(
    df.sort_values('EDI')
      [['STATNAME','DISTNAME','EDI']]
      .head(10)
)
```

```python
# Exercise 37

cols = [
    'TOTPOPULAT',
    'SCHTOT',
    'OVERALL_LI'
]

flags = pd.DataFrame(index=df.index)

for col in cols:

    q1 = np.percentile(df[col],25)
    q3 = np.percentile(df[col],75)

    iqr = q3-q1

    low = q1-1.5*iqr
    high = q3+1.5*iqr

    flags[col] = (
        (df[col]<low) |
        (df[col]>high)
    )

df['OUTLIER_COUNT'] = flags.sum(axis=1)

display(
    df[df['OUTLIER_COUNT']>=2]
    [['STATNAME','DISTNAME','OUTLIER_COUNT']]
)
```

```python
# Exercise 39

df['INFRA_BAND'] = pd.cut(
    df['INFRA_SCORE'],
    bins=3,
    labels=['Low','Medium','High']
)

cross = pd.crosstab(
    df['LIT_BAND'],
    df['INFRA_BAND']
)

display(cross)

plt.imshow(cross)

plt.colorbar()

plt.xticks(range(3),cross.columns)
plt.yticks(range(3),cross.index)

for i in range(cross.shape[0]):
    for j in range(cross.shape[1]):
        plt.text(
            j,
            i,
            cross.iloc[i,j],
            ha='center'
        )

plt.title("Literacy vs Infrastructure")

plt.show()
```

```python
# Exercise 40

simple_avg = df['OVERALL_LI'].mean()

weighted_avg = np.sum(
    df['OVERALL_LI']*df['TOTPOPULAT']
)/np.sum(df['TOTPOPULAT'])

print("Simple Average :",simple_avg)

print("Weighted Average :",weighted_avg)

comparison = (
    df.groupby('STATNAME')
      .apply(
          lambda x: pd.Series({
              'SIMPLE_AVG':x['OVERALL_LI'].mean(),
              'WEIGHTED_AVG':np.sum(
                  x['OVERALL_LI']*x['TOTPOPULAT']
              )/np.sum(x['TOTPOPULAT'])
          })
      )
)

comparison['DIFF'] = abs(
    comparison['SIMPLE_AVG']-
    comparison['WEIGHTED_AVG']
)

comparison = comparison.sort_values(
    'DIFF',
    ascending=False
).head(10)

comparison[['SIMPLE_AVG','WEIGHTED_AVG']].plot(
    kind='bar',
    figsize=(10,5)
)

plt.ylabel("Literacy")

plt.title("Simple vs Weighted Literacy")

plt.show()
```
